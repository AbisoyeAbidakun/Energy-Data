# Dealing with missing data  

 -  The mssing data will be handled by imputing  missing data with the most common unit based on their frequency (i.e. Mode)

```python
# Repacing the missing data with the mode i.e the most common unit
data_fuel['fuel_unit']=data_fuel['fuel_unit'].fillna(data_fuel['fuel_unit'].mode().iloc[0])
```

# Splitting the Data into Numerical and Categorical Data 

```python
# Data Classifier function
# Classifies or categorizes data
def categorizer(dat):
    
#we shall separate the categorical and numeric columns
#segregate training dataset into categorical and numerical
    cat_data = []
    num_data = []
    for i,c in enumerate(dat.dtypes):
        print(i,c)
        if c == object or c == int:
            cat_data.append(dat.iloc[:, i])
        else :
            num_data.append(dat.iloc[:, i])

    cat_data = pd.DataFrame(cat_data).transpose()
    num_data = pd.DataFrame(num_data).transpose()
    
    return [cat_data, num_data]

```

```python
 cat_data, num_data = categorizer(data_fuel)
```

 Converting the Utility Id into categorical variables since it is in actual sense a categorical variable represented numerically and adding it to the categorical variables
